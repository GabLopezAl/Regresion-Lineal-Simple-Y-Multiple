**Importar librerías**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mt
import seaborn as sns
```

**Leer archivo csv**

```python
df_berlin = pd.read_csv("Datos_limpios_Berlin.csv", encoding="MacRoman")
df_berlin.head()
```

```python
# Eliminar columna "Unnamed: 0"
df_berlin = df_berlin.drop(['Unnamed: 0'], axis = 1)
df_berlin.head()
```

```python
# Visualizar los tipos de habitacion que existen
df_berlin["room_type"].unique()
#df_entire_home_apt = df_berlin["room_type"].unique()
#df_entire_home_apt[0]
```

```python
# Convertimos datos tipo str a int o float
from sklearn.preprocessing import LabelEncoder

# Crea una instancia de LabelEncoder
le = LabelEncoder()

# Ajusta el encoder a la columna 'room_type' y la transforma
df_berlin['room_type'] = le.fit_transform(df_berlin["room_type"])

# Muestra los resultados
df_berlin[['room_type']]

#df_berlin["room_type_num"].unique()

```

```python
# Graficar el scatter plot de las variables dependientes e independientes
from turtle import color

sns.scatterplot(x='price', y='room_type', color="blue", data=df_berlin)
sns.scatterplot(x='host_acceptance_rate', y='room_type', color="red", data=df_berlin)
```

```python
# Declaramos nuestras variables dependientes e independientes para nuestra regresion lineal simple
var_indep =  df_berlin[["price"]]
var_dep = df_berlin["room_type"]
```

```python
# Se define model como la regresion lineal
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
type(model)
```

```python
# Entrenamos el modelo con las variables independientes (X) y la variable dependiente (y)
model.fit(X=var_indep, y=var_dep)
```

```python
# Verificamos los coeficientes obtenidos para el modelo ajustado coef-a , intercept-b
model.__dict__
```

**Modelo matematico: y = -0.00290929 + 1.0350962438751357**

```python
# Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R Determinaci
model.score(var_indep,var_dep)
```

```python
# Predecimos los valores de Tipo de habitacion a partir de la var price (nos da las predicciones, numero igual de filas)
y_pred= model.predict(X=df_berlin[['price']])
y_pred
```

```python
# Insertamos la columna de predicciones en el DataFrame, a un lado de la variable dependiente
df_berlin.insert(15, 'Predicciones', y_pred)

# Mostramos las predicciones junto a la variable a predecir
df_berlin[["Predicciones", "room_type"]]
```

```python
# Visualizamos la grafica comparativa entre el total real y el total predecido
sns.scatterplot(x='price', y='room_type', color="blue", data=df_berlin)
sns.scatterplot(x='price', y='Predicciones', color="red", data=df_berlin)
```

```python
# Corroboramos cual es el coeficiente de Determinacion de nuestro modelo
coef_Deter = model.score(X=var_indep, y=var_dep)
coef_Deter
```

```python
# Corroboramos cual es el coeficiente de Correlacion de nuestro modelo
coef_Correl = np.sqrt(coef_Deter)
coef_Correl
```

```python
# Encontramos todas las correlaciones entre las variables
Corr_Factors = df_berlin.select_dtypes(include=['number']).corr()
Corr_Factors
```

```python
# Encontramos el valor absoluto de todas las correlaciones entre las variables
Corr_Factors1 = abs(Corr_Factors)
Corr_Factors1
```

```python
# Graficamos el mapa de color de las coeficientes de correlacion
Heat_Map = sns.heatmap(Corr_Factors1, cmap = 'coolwarm')
```

```python
# Ajustamos el mapa de color de los coeficientes de correlacion
Heat_Map = sns.heatmap(Corr_Factors1, cmap = 'Blues', annot=True, fmt = '.2f')
Heat_Map
```
